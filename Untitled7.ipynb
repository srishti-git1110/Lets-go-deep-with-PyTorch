Goal of the nb is to understand what's actually happening with all these pytorch commands, rather than building a cool classifier with a few lines of code. This understanding is important to get well verse w PyTorch.

```python
from torchvision import models # models contain a lot of pretrained cv models
```

```python
resnet_instance = models.resnet101(pretrained=True) # creating an instance of resnet101 pretrained model, pretrained=True downloads the pretrained weights
                                                    # else we'll just get a randomly initialized resnet101 instance
```

`transforms` from `torchvision` provides us with tools to quickly define basic image preprocessing pipelines

```python
import torch
from torchvision import transforms
preprocess = transforms.Compose([transforms.Resize(256),
                                 transforms.CenterCrop(224),
                                 transforms.ToTensor(),
                                 transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                      std=[0.229, 0.224, 0.225]
                                                      )])
from PIL import Image
img = Image.open('bobby.jpg')
```

```python
img
```

This preprocessing and unsqueezing is the requirement of this network - it's how it expects the inputs (it's how it was trained) so we need to put the input in that std form.

```python
img_prep = preprocess(img)
img_prep.shape
```

```python
batch_t = torch.unsqueeze(img_prep, 0)
batch_t.shape
```

put in evaluation model and get your results!

```python
resnet_instance.eval()
out = resnet_instance(batch_t)
out
```

```python
out.shape
```

```python
_, index = out.max(1)
index
```

```python
percentage = torch.nn.functional.softmax(out, dim=1)[0] * 100
percentage[index[0]].item()
```

```python
[print(x.item()) for x in torch.sort(percentage, descending=True)[0][:5]]
```
